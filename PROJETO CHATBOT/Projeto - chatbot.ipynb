```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_excel('database1.xlsx') #definir df
df
```

```python
#def treino/teste
df_sem_10_ult = df.head(len(df)-10) #parcela de treino
df_10_ult = df.tail(10) #parcerla de teste
```

```python
#definir X e y train
X_train = df_sem_10_ult.drop(['id', 'yr_built', 'floors', 'sqft_lot', 'sqft_living', 'grade', 'date', 'CARO', 'sqft_above', 'sqft_basement', 'yr_renovated', 'view', 'price', 'PPA', 'zscore', 'sqft_living15', 'sqft_lot15', 'waterfront', 'lat', 'long', 
             'zipcode'], axis=1)
y_train = df_sem_10_ult['CARO']

decision_tree = tree.DecisionTreeClassifier(max_depth = 7)
decision_tree.fit(X_train, y_train)

#gerar arvore png
with open("chatbot.dot", 'w') as f: 
     f = tree.export_graphviz(decision_tree,
                              out_file=f,
                              max_depth = 7,
                              impurity = True,
                              feature_names = list(df_sem_10_ult.drop(['id', 'floors', 'sqft_lot', 'sqft_living', 'yr_built', 'grade', 'date', 'CARO', 'sqft_above', 'sqft_basement', 'yr_renovated', 'view', 'price', 'PPA', 'zscore', 'sqft_living15', 'sqft_lot15', 'waterfront', 'lat', 'long', 
             'zipcode'], axis=1)),
                              class_names = ['0', '1', '2'],
                              rounded = True,
                              filled= True )
```

```python
#definir x e y teste
X_teste = df_10_ult.drop(['id', 'grade', 'floors', 'sqft_lot', 'sqft_living', 'yr_built', 'date', 'CARO', 'sqft_above', 'sqft_basement', 'yr_renovated', 'view', 'price', 'PPA', 'zscore', 'sqft_living15', 'sqft_lot15', 'waterfront', 'lat', 'long', 
             'zipcode'], axis=1)
y_teste= df_10_ult['CARO']
```

```python
pred = decision_tree.predict(X_teste)

#precisão
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_teste, pred))
print("\n")
print(classification_report(y_teste, pred))

```
